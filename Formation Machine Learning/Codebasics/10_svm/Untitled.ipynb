```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
iris.target
```

```python
df2=df.copy()
```

```python
df2['target']=iris.target
df2
```

```python
df2['target_name']=df2.target.map(lambda x: iris.target_names[x])
```

```python
df2
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], c=df2.target)
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.legend()
```
